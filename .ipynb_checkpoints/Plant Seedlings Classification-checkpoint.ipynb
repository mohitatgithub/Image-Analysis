### Introduction
The Aarhus University Signal Processing group, in collaboration with University of Southern Denmark, has recently released a dataset containing images of approximately 960 unique plants belonging to 12 species at several growth stages. The challenge was to classify images of these 12 species with high accuracy. Competition & dataset details are given [here](https://www.kaggle.com/c/plant-seedlings-classification).

In this notebook I will discuss some of the data preprocessing steps I used with dataset using openCV followed by a naive approach to solve problem.

```python
import numpy as np
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
%matplotlib inline
```

#### Image Data Preprocessing

```python
#Reading training data id & labels in dataframe
data = []
for folder in sorted(os.listdir('/media/mohit/SSD/Datasets/Plant Seedlings Classification/train/')):
    for file in sorted(os.listdir('/media/mohit/SSD/Datasets/Plant Seedlings Classification/train/'+folder)):
        data.append((folder, os.path.splitext(file)[0]))

train_df = pd.DataFrame(data, columns=['Class', 'ImageId'])
train_df.head()
```

```python
#Reading training data id & labels in dataframe
data = []
for file in sorted(os.listdir('/media/mohit/SSD/Datasets/Plant Seedlings Classification/test/')):
        data.append((os.path.splitext(file)[0]))

test_df = pd.DataFrame(data, columns=['ImageId'])
test_df.head()
```

```python
print ('The train data has {} images.'.format(train_df.shape[0]))
print ('The test data has {} images.'.format(test_df.shape[0]))
```

```python
#Checking data distribution

label_counts = train_df.Class.value_counts()
plt.figure(figsize = (12,6))
sns.barplot(label_counts.index, label_counts.values, alpha = 0.9)
plt.xticks(rotation = 'vertical')
plt.xlabel('Image Labels', fontsize =12)
plt.ylabel('Counts', fontsize = 12)
plt.show()
```

There are 12 classes of seedlings, but number of seedlings in each class is between 221 to 654 images for each class. 

```python
# function to read images as arrays
def read_image(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (256,256))
    return img
```

```python
#Creating copy of all training images in seprate folder
import shutil

train_path = '/media/mohit/SSD/Datasets/Plant Seedlings Classification/train/'
test_path = '/media/mohit/SSD/Datasets/Plant Seedlings Classification/test/'

train_all = os.makedirs(os.path.dirname(train_path+'/train_all/'), exist_ok=True)

for folder in sorted(os.listdir(train_path)):
    for file in sorted(os.listdir(train_path+folder)):
        shutil.copy(train_path+folder+'/'+file, train_path+'train_all/')
```

```python
#Reading all images in a python lists
train_data = []
test_data = []
train_labels = train_df['Class'].values

for img in tqdm(train_df['ImageId'].values):
    train_data.append(read_image(train_path+'train_all/' + '{}.png'.format(img)))
    
for img in tqdm(test_df['ImageId'].values):
    test_data.append(read_image(test_path + '{}.png'.format(img)))
```

```python
#Normalizing Images
x_train = np.array(train_data, np.float32) / 255.
x_test = np.array(test_data, np.float32) / 255.
```

```python
#Lets check some images
def show_images(ix):
    image_train = read_image(train_path + 'train_all/' + train_df.ImageId[ix] + '.png')
    image_test = read_image(test_path + test_df.ImageId[ix] + '.png')
    
    pair = np.concatenate((image_train, image_test), axis=1)
    
    plt.figure(figsize = (4,4))
    plt.imshow(pair)
    
#First 4 images in train and test set
for idx in range(4):
    show_images(idx)
```

```python
##Lets see the images with their respective labels
plt.rc('axes', grid=False)

_, axs = plt.subplots(3,3, sharex = 'col', sharey='row', figsize = (7,7))
axs = axs.ravel()

#Lets see first 8 images
for i, (image_name, label) in enumerate(zip(train_df.ImageId, train_df.Class)):
    if i <= 8:
        img = read_image(train_path+'/train_all/' + image_name + '.png')
        axs[i].imshow(img)
        axs[i].set_title('{} - {}'.format(image_name, label))
    else:
        break
```

### Image Classification using CNN Architectures

#### 1. Image Classification using Naive approach

```python
#Lets train our first model using keras

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Convolution2D, MaxPooling2D
from keras.callbacks import EarlyStopping
from keras.utils import to_categorical
```

```python
#Encoding class as numeric value between 1-12
label_list = train_df['Class'].tolist()
Y_train = {k:v+1 for v,k in enumerate(set(label_list))}
y_train = [Y_train[k] for k in label_list]      

#print(set(y_train))
y_train = to_categorical(y_train)
```

```python
##CNN Architechture

model = Sequential()
model.add(Convolution2D(32, (3,3), activation='relu', padding='same',input_shape = (64,64,3))) 
model.add(Convolution2D(32, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Convolution2D(64, (3,3), activation='relu', padding='same'))
model.add(Convolution2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Convolution2D(128, (3,3), activation='relu', padding='same'))
model.add(Convolution2D(128, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(y_train.shape[1], activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
early_stops = EarlyStopping(patience=3, monitor='val_acc')
```

```python
model.fit(x_train, y_train, batch_size=5, epochs=11, validation_split=0.3)
```

```python
#Prediction
predictions = model.predict(x_test)
predictions = np.argmax(predictions, axis= 1)
```

```python
#Getting predicted labels
y_maps = dict()
y_maps = {v:k for k, v in Y_train.items()}
pred_labels = [y_maps[k] for k in predictions]
```

```python
##Lets see what our classifier predicts on test images

# top 5 predictions
for i in range(5):
    print('I see this product is {}'.format(pred_labels[i]))
    plt.imshow(read_image(test_path +'{}.png'.format(test_df.ImageId[i])))
    plt.show()
```
